## Define Methods for preprocessing data

```python
def outlier_truncation(x, factor=1.5):
    '''
    Identifies outlier values based on the inter-quartile range IQR. 
    Corresponding outliers are truncated and set to a contant value equal to the IQR
    times a factor, which, following Tuckey's rule, we set to 1.5 by default
    
        Parameters:
            x (Pandas Series): A data frame column to scan for outliers
            factor (float): An outlier is a value this many times the IQR above q3/below q1
            
        Returns:
            Adjusted variable in which outliers are truncated
    '''
    x_new = x.copy()
    
    # Calculate IQR
    IQR = x.quantile(0.75) - x.quantile(0.25) 
    
    # Define upper/lower bound
    upper = x.quantile(0.75) + factor*IQR
    lower = x.quantile(0.25) - factor*IQR
    
    # Truncation
    x_new[x < lower] = lower
    x_new[x > upper] = upper
    
    return x_new

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

def add_age(df):
    df['age'] = df['user_dob'].apply(calculate_age)
    df['age'] = df['age'].astype(np.float32)
    return df

def add_dob_missing(df):
    df['dob_missing'] = df['user_dob'].apply(lambda x: False if pd.isnull(x) else True)
    return df

def transform_columns(df_known, known_data:bool):
    # change object variables to datatype category
    df_known['item_size'] = df_known['item_size'].astype('category')
    df_known['item_color'] = df_known['item_color'].astype('category')
    df_known['user_title'] = df_known['user_title'].astype('category')
    df_known['user_state'] = df_known['user_state'].astype('category')
    # change all numeric variables from float64 to float32 to reduce memory consumption
    df_known['item_price'] = df_known['item_price'].astype(np.float32)
    df_known['item_price'] = df_known['item_price'].apply(lambda x:("%.2f" % round(x, 2)))
    df_known['item_price'] = df_known['item_price'].astype(np.float32)
    df_known['brand_id'] = df_known['brand_id'].astype(np.int32)
    df_known['user_id'] = df_known['user_id'].astype(np.int32)
    df_known['item_id'] = df_known['item_id'].astype(np.int32)
    if known_data:
    # since the feature return has only two values, we convert it to boolean
        df_known['return'] = df_known['return'].astype('bool')
    # transform all dates to the datetime datatype
    df_known['order_date'] = df_known['order_date'].astype('datetime64[ns]')
    df_known['delivery_date'] = df_known['delivery_date'].astype('datetime64[ns]')
    df_known['user_dob'] = df_known['user_dob'].astype('datetime64[ns]')
    df_known['user_reg_date'] = df_known['user_reg_date'].astype('datetime64[ns]')
    return df_known

def add_delivery_days(df):
    df['delivery_days'] = (df['delivery_date'] - df['order_date']).dt.days
    df['delivery_days'] = df['delivery_days'].astype(np.float32)
    return df 

def remove_delivery_days_outliers(df):
    df['delivery_days'] = outlier_truncation(df['delivery_days']) 
    return df

def add_delivery_date_missing(df):
    df['delivery_date_missing'] = df['delivery_date'].apply(lambda x: False if pd.isnull(x) else True)
    return df

def add_delivery_date_1994_marker(df):
    df['delivery_date_1994'] = df['delivery_date'].apply(lambda x: True if x.year < 2016 else False)
    return df

def set_delivery_date_1994_to_nan(df):
    df.loc[df['delivery_date'].dt.year < 2016,['delivery_days']] = np.nan
    return df

def add_been_member_for(df):
    df['been_member_for'] = (df['order_date']-df['user_reg_date'] ).dt.days
    df['been_member_for'] = df['been_member_for'].astype(np.int32)
    return df

def add_member_category(df):
    labels = ['fresh_member', 'new_member', 'member', 'old_member']
    #cut_bins = [-5, 5, 120, 491, 200000]
    cut_bins = [-5, 150, 300, 450, 1000]
    df['member'] = pd.cut(df['been_member_for'], bins=cut_bins, labels=labels)
    return df

def truncate_item_price_outliers(df):
    df['item_price'] = outlier_truncation(df['item_price'])
    return df

def add_brand_id_count(df):
    df['brand_id_count'] = df['brand_id'].apply(lambda x: (df['brand_id'] == x).sum())
    df['brand_id_count'] = df['brand_id_count'].astype(np.float32)
    return df

def add_item_id_count(df):
    df['item_id_count'] = df['item_id'].apply(lambda x: (df['item_id'] == x).sum())
    df['item_id_count'] = df['item_id_count'].astype(np.float32)
    return df

def add_item_color_count(df):
    # In case the alphabet are sometimes lowercase and sometimes uppercase
    df['item_color'] = df['item_color'].str.lower()

    # Correct some mistakes
    df.loc[df['item_color'] == 'blau','item_color'] = "blue"
    df.loc[df['item_color'] == 'brwon','item_color'] = "brown"
    df.loc[df['item_color'] == 'oliv','item_color'] = "olive"
    df['item_color_count'] = df['item_color'].apply(lambda x: (df['item_color'] == x).sum())
    df['item_color_count'] = df['item_color_count'].astype(np.float32)
    return df

def truncate_age_outliers(df):
    df['age'] = outlier_truncation(df['age'])
    return df

def cut_age_outliers(df):
    df.loc[df['age'] > 95,'age'] = np.nan
    return df

def convert_item_sizes(df):
    sizes_dict = {
        '84': 'xxs',
        '104': 's',
        '110': 's',
        '116': 's',
        '122': 'm',
        '128': 'm',
        '134': 'l',
        '140': 'l',
        '148': 'xl',
        '152': 'xl',
        '164': 'xxl',
        '170': 'xxl',
        '176': 'xxxl',
        '18': 'xs',
        '19': 's',
        '20': 's',
        '21': 'm',
        '22': 'm',
        '23': 'l',
        '24':  'xl',
        '25': 'xs',
        '26': 's',
        '27': 's',
        '28': 'm',
        '29': 'm', 
        '30': 'l',
        '31': 'l',
        '32': 'xl',
        '33': 'xxl',
        '34': 'xxs',
        '35': 'xs',
        '36': 'xs',
        '36+': 's',
        '37': 's',
        '37+': 's',
        '38': 's',
        '38+': 's',
        '39': 'm',
        '39+': 'm',
        '40': 'm',
        '40+': 'm',
        '41': 'm',
        '41+' : 'm',
        '42': 'l',
        '42+': 'l',
        '43': 'l',
        '43+': 'l',
        '44': 'l',
        '44+' : 'xl',
        '45' : 'xl',
        '45+': 'xl',
        '46': 'xl',
        '46+' : 'xl',
        '47' : 'xl',
        '48': 'xl',
        '49': 'xl',
        '50': 'xxl',
        '52': 'xxl',
        '54': 'xxl',
        '56': 'xxl',
        '58': 'xxl',
        0: 'xxs',
        '1': 'xxs',
        '2': 'xxs',
        '2+': 'xxs',
        '3' : 'xxs',
        '3+': 'xs',
        '4':  'xs',
        '4+': 'xs',
        '5': 'xs',
        '5+':'xs',
        '6':'s',
        '6+':'s',
        '7':'s',
        '7+':'m',
        '8':'m',
        '8+':'m',
        '9': 'l',
        '9+': 'l',
        '10': 'l',
        '10+': 'xl',
        '11': 'xl',
        '11+': 'xl',
        '12': 'xl',
        '12+': 'xxl',
        '13': 'xxl',
        '14': 'xxl',
        36: 'xxs',
        38: 'xs',
        40: 's',
        42: 'm',
        44: 'l',
        46: 'xl',
        48: 'xxl',
        '3132': 'xxs',
        '3332': 'xs',
        '3432': 'xs',
        '3632': 's',
        '3832': 'm',
        '3634': 'l',
        '3834': 'xl',
        '4032': 'xl',
        '4034': 'xxl',
        '4232': 'xxxl',
        '80': 'xs',
        '85': 's',
        '90': 'm',
        '95': 'l',
        '100': 'xl',
        '105': 'xxl'
    }
#     sizes_dict = {
#         '84': 'Kids',
#         '104': 'Kids',
#         '110': 'Kids',
#         '116': 'Kids',
#         '122': 'Kids',
#         '128': 'Kids',
#         '134': 'Kids',
#         '140': 'Kids',
#         '148': 'Kids',
#         '152': 'Kids',
#         '164': 'Kids',
#         '170': 'Kids',
#         '176': 'Kids',
#         '18': '18-33',
#         '19': '18-33',
#         '20': '18-33',
#         '21': '18-33',
#         '22': '18-33',
#         '23': '18-33',
#         '24':  '18-33',
#         '25': '18-33',
#         '26': '18-33',
#         '27': '18-33',
#         '28': '18-33',
#         '29': '18-33', 
#         '30': '18-33',
#         '31': '18-33',
#         '32': '18-33',
#         '33': '18-33',
            
#         '34': '34-58',
#         '35': '34-58',
#         '36': '34-58',
#         '36+': '34-58',
#         '37': '34-58',
#         '37+': '34-58',
#         '38': '34-58',
#         '38+': '34-58',
#         '39': '34-58',
#         '39+': '34-58',
#         '40': '34-58',
#         '40+': '34-58',
#         '41': '34-58',
#         '41+' : '34-58',
#         '42': '34-58',
#         '42+': '34-58',
#         '43': '34-58',
#         '43+': '34-58',
#         '44': '34-58',
#         '44+' : '34-58',
#         '45' : '34-58',
#         '45+': '34-58',
#         '46': '34-58',
#         '46+' : '34-58',
#         '47' : '34-58',
#         '48': '34-58',
#         '49': '34-58',
#         '50': '34-58',
#         '52': '34-58',
#         '54': '34-58',
#         '56': '34-58',
#         '58': '34-58',
            
#         0: '0-14',
#         '1': '0-14',
#         '2': '0-14',
#         '2+': '0-14',
#         '3' : '0-14',
#         '3+': '0-14',
#         '4':  '0-14',
#         '4+': '0-14',
#         '5': '0-14',
#         '5+':'0-14',
#         '6':'0-14',
#         '6+':'0-14',
#         '7':'0-14',
#         '7+':'0-14',
#         '8':'0-14',
#         '8+':'0-14',
#         '9': '0-14',
#         '9+': '0-14',
#         '10': '0-14',
#         '10+': '0-14',
#         '11': '0-14',
#         '11+': '0-14',
#         '12': '0-14',
#         '12+': '0-14',
#         '13': '0-14',
#         '14': '0-14',
            
#         36: '36-48',
#         38: '36-48',
#         40: '36-48',
#         42: '36-48',
#         44: '36-48',
#         46: '36-48',
#         48: '36-48',
            
#         '3132': '3132-4232',
#         '3332': '3132-4232',
#         '3432': '3132-4232',
#         '3632': '3132-4232',
#         '3832': '3132-4232',
#         '3634': '3132-4232',
#         '3834': '3132-4232',
#         '4032': '3132-4232',
#         '4034': '3132-4232',
#         '4232': '3132-4232',
            
#         '80': '80-105',
#         '85': '80-105',
#         '90': '80-105',
#         '95': '80-105',
#         '100': '80-105',
#         '105': '80-105',
            
#         'xxs': 'xxs-xxl',
#         'xs': 'xxs-xxl',
#         's': 'xxs-xxl',
#         'm': 'xxs-xxl',
#         'l': 'xxs-xxl',
#         'xl': 'xxs-xxl',
#         'xxl': 'xxs-xxl',
#         'xxxl': 'xxs-xxl'
#     }

    df['item_size'] = df['item_size'].str.lower()
    df.replace({'item_size':sizes_dict}, inplace = True)
    df['item_size'] = df['item_size'].astype('category')
    return df
```

## Information Gain

```python
for c in df.select_dtypes(include='float32').columns:
    df[c] = df[c].astype('float64')
for c in df.select_dtypes(include='int32').columns:
    df[c] = df[c].astype('int64')
```

```python
class filter_binary_target:
    def __init__(self, df, target):
        self.target = target
        self.data_head = df.head()

    def auto_filter_binary_target(self):
        print('Data must be in a clean pandas DataFrame. Categorical variables must be of data type bool or category. Continuous variables must be int64 or float64.')
        data_no_target = df.drop(columns=self.target)
        columns = ['Data Type', 'Metric', 'Score']
        index = data_no_target.columns
        result = pd.DataFrame(index=index, columns=columns)

        for col in data_no_target:
            if data_no_target.dtypes[col] == 'bool' or data_no_target.dtypes[col].name == 'category':
                result.loc[col, 'Data Type'] = "discrete"
                result.loc[col, 'Metric'] = "IV"
                result.loc[col, 'Score'] = self.IV_binary_target(feature=col)

            if data_no_target.dtypes[col] == 'int64' or data_no_target.dtypes[col] == 'float64':
                result.loc[col, 'Data Type'] = "continuous"
                result.loc[col, 'Metric'] = "Fisher"
                result.loc[col, 'Score'] = self.fisher_binary_target(feature=col)
                
        return result

    def IV_binary_target(self, feature):  # same code as used above
        data = pd.DataFrame()
    
        data['Count'] = df[feature].value_counts()
        data['Bad'] = df.groupby([feature])[self.target].sum()
        data['Good'] = data['Count'] - data['Bad']
    
        data["Distribution Bad"] = data["Bad"] / data["Bad"].sum()
        data["Distribution Good"] = data["Good"] / data["Good"].sum()
    
        data['WOE'] = np.log(data["Distribution Good"] / data["Distribution Bad"])
        data.replace({"WOE": {np.inf: 0, -np.inf: 0}})

        data["IV"] = data["WOE"] * (data["Distribution Good"] - data["Distribution Bad"])

        iv = data["IV"].sum()

        return iv

    def fisher_binary_target(self, feature):
        mu_0 = df.groupby(df[self.target])[feature].mean()[0]
        mu_1 = df.groupby(df[self.target])[feature].mean()[1]
        var_0 = df.groupby(df[self.target])[feature].var()[0]
        var_1 = df.groupby(df[self.target])[feature].var()[1]

        num = abs(mu_0 - mu_1)
        den = (var_0 + var_1) ** 0.5
        score = num/den
    
        return score

    def pearson(self, feature):  # since our target is binary, we actually don't need this. However, if you would like to expand this class, you can use this code
        mean_feature = df[feature].mean()
        mean_target = df[self.target].mean()
        num = ((df[feature] - mean_feature)*(df[self.target] - mean_target)).sum()
        den = (((df[feature] - mean_feature)**2).sum() * ((df[self.target] - mean_target)**2).sum()) ** .5
        rho = num/den
        return rho
```

```python
# I will use OneHotEncoder for the categorical features
categorical_transformer = OneHotEncoder(handle_unknown='error', drop='first')
# and a MinMaxScaler for the boolean features
identity_transformer = MinMaxScaler()

# The numeric features I will first impute, then scale and normalize at the end.
# Since there are multiple imputing strategies, I define the variablenumeric_imputer_strategy,
# which can be set to different values to inspect their influence on the model.
# In order to check if centering the values before scaling influences the performance,
# I created the variable numeric_standard_scaler_mean.

numeric_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy=numeric_imputer_strategy, add_indicator=True)),
        ('scaler', StandardScaler(with_mean=numeric_standard_scaler_mean)),
        ('normalizer', Normalizer())
    ]
)

```

## Cost sensitivity

Even though `Light GBM` performed best in terms of predicting unseen data, I would still want to keep all models to evaluate which one would  result in the lowest cost. For this purpose, I created the following dictionary, which would make the execution of repetitive tasks for all models much easier and clearer to understand.

```python
models = {'Logistic Regression': clf_lr, 'XGB': clf_xgb, 'Random Forest': clf_rf, 'Light GBM': clf_lgb}
```

### Get probaility predictions for each model

First, I create the dictionary `pred_proba_dict`, in which I am going to keep the probability predictions of each model, so that I don't have to calculate them every time I need them.

```python
pred_proba_dict = dict()
for model_name, model in models.items():
    pred_proba = model.predict_proba(X_test)[:,1]
    pred_proba_dict[model_name] = pred_proba
```

### Calculate bayesian threshold

Next, I calculate average costs for the items with the assumption that all of them were classified as False Positives (FP) or as False Negatives (FN). Then I will use these values to create the cost matrix.

```python
fn_costs = X_test.apply(lambda row: (0.5*5*(3+0.1*row['item_price'])), axis= 1)
fn_costs_avg = np.mean(fn_costs)
fp_costs = X_test.apply(lambda row: (0.5*row['item_price']), axis= 1)
fp_costs_avg = np.mean(fp_costs)
tp_costs = 0
tn_costs = 0
print('Avg. FN cost: ' + str(fn_costs_avg))
print('Avg. FP cost: ' + str(fp_costs_avg))
```

```python
cost_matrix = pd.DataFrame({'actual_keep(0)': [tp_costs, fp_costs_avg], 'actual_return(1)': [fn_costs_avg, tn_costs]}, index= ['predicted_keep(0)', 'predicted_return(1)'])
cost_matrix
```

With the help of the cost matrix, I will calculate the bayes threshold

```python
# Bayes optimal threshold
def calculate_threshold_bayes(cost_matrix):
    threshold_bayes = (cost_matrix.iloc[0,1]               # C(b,G)
                   /(cost_matrix.iloc[1,0]             # C(b,G)
                     +cost_matrix.iloc[0,1])).round(5) # C(g,B)
    return threshold_bayes
```

```python
threshold_bayes = calculate_threshold_bayes(cost_matrix)
threshold_bayes
```

### Calculate error cost per model sum with threshold = 0,5

Here I calculate the sum of the cost for each model with the default threshold = 0,5

```python
def calc_all_error_cost_per_model(y_probabilities, y_test, cutoff):
    pred = np.where(y_probabilities >= cutoff, 1, 0)                
    #cmat = metrics.confusion_matrix(y_test, pred)                      
    false_positives_all = np.logical_and(y_test != pred, pred == 1)
    false_positives = false_positives_all[false_positives_all].index
    false_negatives_all = np.logical_and(y_test != pred, pred == 0)
    false_negatives = false_negatives_all[false_negatives_all].index
    
    df = X_test.copy()
    df['cost_fn'] = df.apply(lambda row: (0.5*5*(3+0.1*row['item_price'])) if row.name in false_negatives else 0, axis= 1)
    df['cost_fp'] = df.apply(lambda row: (0.5*row['item_price']) if row.name in false_positives else 0, axis= 1)
    
    #print(df.head())
    cost_fn = df['cost_fn'].sum()
    cost_fp = df['cost_fp'].sum()
    cost = cost_fn + cost_fp
    return cost
```

```python
default_costs_dict = dict()
for model_name, model in models.items():
    overall_costs = calc_all_error_cost_per_model(pred_proba_dict[model_name], y_test, 0.5)
    default_costs_dict[model_name] = overall_costs
```

```python
default_costs_dict
```

### Calculate error cost per model mean with threshold = 0,5

Here I calculate the mean of the FP and FN costs for each model with the default threshold = 0,5, in order to compute the cost matrix afterwards.

```python
def calc_error_cost_per_model_mean(y_probabilities, y_test, cutoff):
    pred = np.where(y_probabilities >= cutoff, 1, 0)                    # convert class probabilities into class predictions
    #cmat = metrics.confusion_matrix(y_test, pred)                       # compute corresponding confusion matrix
    false_positives_all = np.logical_and(y_test != pred, pred == 1)
    false_positives = false_positives_all[false_positives_all].index
    false_negatives_all = np.logical_and(y_test != pred, pred == 0)
    false_negatives = false_negatives_all[false_negatives_all].index
    
    df = X_test.copy()
    df['cost'] = 0
    df['cost_fn'] = df.apply(lambda row: (0.5*5*(3+0.1*row['item_price'])) if row.name in false_negatives else 0, axis= 1)
    df['cost_fp'] = df.apply(lambda row: (0.5*row['item_price']) if row.name in false_positives else 0, axis= 1)
    
    cost_fn = np.mean(df['cost_fn'])
    cost_fp = np.mean(df['cost_fp'])
    
    return {'cutoff': cutoff, 'fn_costs': cost_fn, 'fp_costs': cost_fp}
```

```python
mean_costs_per_model_dict = dict()
for model_name, model in models.items():
    overall_costs = calc_error_cost_per_model_mean(pred_proba_dict[model_name], y_test, 0.5)
    mean_costs_per_model_dict[model_name] = overall_costs
```

```python
cost_matrix_mean_dict = dict()
for model_name, model in models.items():
    tp_costs = 0
    tn_costs = 0
    fp_costs = mean_costs_per_model_dict[model_name]['fp_costs']
    fn_costs = mean_costs_per_model_dict[model_name]['fn_costs']
    cost_matrix = pd.DataFrame({'actual_keep(0)': [tp_costs, fp_costs], 'actual_return(1)': [fn_costs, tn_costs]}, index= ['predicted_keep(0)', 'predicted_return(1)'])
    cost_matrix_mean_dict[model_name] = cost_matrix
    display(cost_matrix)
```

### Calculate bayesian threshold for each model

With the new cost matrix for each model, I calculate the bayesian threshold for each model.

```python
threshold_bayes_dict_mean = dict()
for model_name, model in models.items():
    threshold_bayes_dict_mean[model_name] = calculate_threshold_bayes(cost_matrix_mean_dict[model_name])
threshold_bayes_dict_mean
```

### Calculate error cost per model mean with  bayesian threshold

In this step, I calculate the mean of the FP and FN costs for each model with the bayesian threshold for the respective model.

```python
bayesian_costs_per_model_dict = dict()
for model_name, model in models.items():
    overall_costs = calc_error_cost_per_model_mean(pred_proba_dict[model_name], y_test, threshold_bayes_dict_mean[model_name])
    bayesian_costs_per_model_dict[model_name] = overall_costs
bayesian_costs_per_model_dict
```

Then, the percentage of observations that are now predicted with the label 1 are presented by model.

```python
pred_default_dict = dict()
pred_bayes_dict = dict()
for model_name, model in models.items():
    print(model_name)
    pred_proba = pred_proba_dict[model_name]
    threshold_bayes = threshold_bayes_dict_mean[model_name]
    pred_default = np.where(pred_proba >= 0.5, 1, 0)
    pred_default_dict[model_name] = pred_default
    pred_bayes= np.where(pred_proba >= threshold_bayes_dict_mean[model_name], 1, 0) 
    pred_bayes_dict[model_name] = pred_bayes
    print('Default threshold: ' + str(np.mean(pred_default)), 'Bayes threshold: ' + str(np.mean(pred_bayes)))
```

## Evaluating cost-sensitive classifiers

Next, I will compare the default and the bayes threshold.

### Accuracy

Interestingly, when using the bayesian threshold, the accuracy of the model drops. Nevertheless, this is not dramatic, as the costs also drop and after all this is what we are trying to achieve! In the  following function, the cut off for each of the thresholds is calculated and presented in a plot.

```python
# Calculating fpr and tpr manually at a given cutoff
# We will use this function to plot markers at distinct point across our line graph to see where the different cutoffs are along the graph
def tpr_fpr_calc(cutoff, yhat_prob, y_true):

    # temp variable giving distinct predictions based on the cutoff given as an input
    temp = (yhat_prob >= cutoff).astype(bool)

    # Create confusion matrix for this prediction
    cmat = confusion_matrix(y_true, temp)
    
    # Calculate FPR
    fpr = cmat[0,1] / (cmat[0,1] + cmat[0,0])

    # Calculate TPR
    tpr = cmat[1,1] / (cmat[1,1] + cmat[1,0])

    return tpr, fpr
```

If we take a look at the cost matrix, we see that it is better if we have higher Sensitivity.

For the default cutoff both Sensitivity and Specifity for each model is around 60%. But with the bayes threshold, the sensitivity increases a lot for each model and almost reaches 100%, whereas the Specifity drops to around 20% for each model.

## Empirical Thresholding

Next, I will perform Empirical Thresholding to see if there is another threshold, that would result in lower costs for my models. Here, the goal is not to find an auc-maximizing cutoff, but a cost-minimizing one. Therefore, a cross-validation approach is performed with an average over all cutoffs with the lowest error-cost for each fold.

```python
# To use cross-validation we refer back to this function we defined in the model assessment tutorial
from sklearn.utils import shuffle
def cv_split(seq, splits):
    return (seq[i::splits] for i in range(splits))

# Define function for threshold tuning
def lowest_error_cutoff(model, k, X_train_df, X_test_df, y_train_true, y_test_true, bayes_cost_matrix, y_test_probabilities):
    cutoffs_total = [] # save best cutoff from each split here, so we can average this cutoff later

  # Create a list of the index of our full data set as input, and k for the amount of splits
    cv_split_list = list(cv_split(shuffle(X_train_df.index.values.tolist(), random_state=888), k)) 
    plt.figure()
    for fold, i in enumerate(cv_split_list): # for each split 

        temp = X_train_df.index.isin(i) # Create temp variable for data to exclude            
        X_train_cv, y_train_cv = X_train_df.iloc[~temp], y_train_true[~temp] # Build model on all but hold out data
        X_test_cv, y_test_cv =  X_train_df.iloc[temp], y_train_true[temp] # Test on hold out data later
        model.fit(X_train_cv, y_train_cv.ravel()) # Train model
        pred_temp_prob = model.predict_proba(X_test_cv)[:, 1] # Calculate class probabilities for current split

    # Define empty variable to save all cutoffs and corresponding errors. This way we can extract the lowest one for this split later
    cutoffs_cv = []
    error_cost_cv = []

    for j in  np.arange(0, 1, 0.001): # np.arrange(from, to, step-size)
        pred_temp = (pred_temp_prob >= j).astype(bool) # convert probabilities to class predictions
        cutoffs_cv.append(j) # save corresponding cutoff
        temp_cutoff_cost = calc_error_cost(pred_temp, y_test_cv, j, bayes_cost_matrix) # calculate cost and cutoff using our previously function
        error_cost_cv.append(temp_cutoff_cost[1]) # save error for the given cutoff
      
    temp_df = pd.DataFrame({"cutoffs":cutoffs_cv, "error_cost": error_cost_cv}).sort_values("error_cost")  # merge cutoffs and error cost, sort from lowest to highest by error_cost
    print("This is the fold number {}".format(fold+1)," of the cross-validation")
    print("The lowest error on the holdout data was achieved with the following cut-off: {}".format(temp_df.iloc[0].values[0]))
    print("The corresponding error cost is: {}".format(temp_df.iloc[0].values[1]))
    print("")
    cutoffs_total.append(temp_df.iloc[0, 0]) # save the lowest cutoff of this split to the list so we can average it at the end
    plt.plot(temp_df.cutoffs, temp_df.error_cost) # plot cutoffs and corresponding errors
   
    temp_cutoff_cost= calc_error_cost(y_test_probabilities, y_test_true, np.mean(cutoffs_total).round(5), bayes_cost_matrix) # np.mean(cutoffs_total) is the resulting best cutoff from the cross validation. We calculate the corresponding error-cost on the test data set
    plt.plot(temp_cutoff_cost[0], temp_cutoff_cost[1], marker='x', markersize=20, alpha=1, color='black', label="Average cutoff performance on test data") # plot cutoff and corresponding error on test set
    plt.title("Error Cost on Holdout Data by Threshold")
    plt.ylabel("Error Cost")
    plt.xlabel("Threshold") 

    print("The best threshold according to cross-validated threshold tuning is: {}".format(temp_cutoff_cost[0])) 
    print("On the test set, it has the following error cost: {}".format(temp_cutoff_cost[1]))
    return temp_cutoff_cost
```

## The MetaCost algorithm

Lastly, I will try the MetaCost algorithm. In the first step of the algorithm, I apply the minimal bayes cutoff and use it to label our data. With the probability predictions for the test set for this model, I apply the Bayes minimal cutoff to it. Then this vector is used as the new y_train. Next, another model is trained on the data including the new labels then predict the output of the test set. As the first model, I use Logistic Regression. For the case of Logistic Regression itself, I chose to use XGB as the trained model to label the data.

```python
def MetaCost(X_train, X_test, y_train, relabel_model, final_model, optimal_cutoff):
  # Use model with good calibration to predict the new labels and create probability predictions
  temp_pred_proba_train = relabel_model.fit(X_train, y_train.ravel()).predict_proba(X_train)[:,1]

  # Convert the probability predictions to class predictions using Bayes Decision Theory
  temp_pred_class_train = np.where(temp_pred_proba_train >= optimal_cutoff, 1, 0) # These are the new class labels

  # Fit a potent model to the test set, replacing our usual y_train with resulting class predictions from the last step
  final_model.fit(X_train, temp_pred_class_train.ravel())

  # Get predictions for the test set
  temp_pred_proba_test_cs = final_model.predict_proba(X_test)[:,1]

  return temp_pred_proba_test_cs
```

```python
cutoff_cost_MC_dict = dict()
cmat_mc_dict = dict()
for model_name, model in models.items():
    if model_name == 'Logistic Regression':
        model_1 = models['XGB']
    else:
        model_1 = models['Logistic Regression']
    model_2 = model
    # Apply function 
    pred_proba_MC = MetaCost(X_train, X_test, y_train, model_1, model_2, threshold_bayes_dict_mean[model_name])

    # Train model with original labels and predict the test set for comparison
    pred_proba_2 = model_2.fit(X_train, y_train.ravel()).predict_proba(X_test)[:,1];

    # Save corresponding confusion matrix for evaluation
    cmat_MC = metrics.confusion_matrix(y_test, np.where(pred_proba_MC >=  0.5, 1, 0))
    cmat_mc_dict[model_name] = cmat_MC

    # Evaluate by calcuating the error-cost
    cutoff_cost_MC = calc_error_cost(pred_proba_MC, y_test, 0.5, cost_matrix_mean_dict[model_name])
    cutoff_cost_MC_dict[model_name] = cutoff_cost_MC
    cutoff_cost_2 = calc_error_cost(pred_proba_2, y_test, 0.5, cost_matrix_mean_dict[model_name])

    print("The error-cost for the regular model of {} is: {}".format(model_name, cutoff_cost_2[1]))
    print("The error-cost for the MetaCost approach of {} is: {}".format(model_name, cutoff_cost_MC[1]))
```
